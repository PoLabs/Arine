```python
import numpy as np
import pandas as pd
import warnings
from IPython.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
warnings.filterwarnings('ignore')
df_ad = pd.read_csv('/home/polabs2/Code/Arine/admissions.csv')
print(df_ad.columns)
df_ad['admit_date_new'] = pd.to_datetime(df_ad['admit_date'])

from pandas_profiling import ProfileReport
warnings.filterwarnings('ignore')
profile = ProfileReport(df_ad, title="Profiling Report")
profile
```

```python
%matplotlib inline
from matplotlib import pyplot
df_ad['admit_date_new'] = pd.to_datetime(df_ad['admit_date'])
df_timeline = df_ad.groupby(['claim_number'], as_index=False).agg({'patient_id':'first', 'admit_date_new':'first'}).sort_values(by=['admit_date_new'])
df_timeline['month_year'] = pd.to_datetime(df_timeline['admit_date_new']).dt.to_period('M')
df_timeline = df_timeline.groupby(['month_year'], as_index=False).agg({'patient_id':'count'}) # pd.Series.nunique
df_timeline
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
df_timeline.set_index('month_year')
plt.plot_date(df_timeline['month_year'], df_timeline['patient_id'])
plt.gcf().autofmt_xdate()
plt.show()
```

## Part A: We see a clear drop off in inpatient admissions starting in April 2020
#### To my knowledge, elective surgeries were for the most part canceled or postponed during the COVID-19 pandemic - which would directly impact inpt admissions. 
#### Although the pandemic really 'hit' the US in March, many surgeries were probably already planned for March which is why the big drop was in April. It also took some time for the extend of the pandemic impact to be acknowledged. 
#### We see a slight climb after April once the infection rate was better controlled and transmission precautions were better/more widely implemented.
